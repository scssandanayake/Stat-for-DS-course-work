```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df = pd.read_csv("/content/SriLanka_Weather_Dataset.csv")
```

```python
selected_columns = df[['rain_sum', 'temperature_2m_mean', 'city']]
```

```python
ra_df = selected_columns[selected_columns['city'] == 'Ratnapura']
```

```python
ga_df = selected_columns[selected_columns['city'] == 'Galle']
```

```python
ra_df_rain = ra_df['rain_sum'].to_frame()
ga_df_rain = ga_df['rain_sum'].to_frame()
```

```python
ra_df_rain.reset_index(inplace = True)
ga_df_rain.reset_index(inplace = True)

df_new2 = pd.DataFrame()
df_new2['rathnapura_rain'] = ra_df_rain['rain_sum']
df_new2['galle_rain'] = ga_df_rain['rain_sum']
print(df_new2)
```
